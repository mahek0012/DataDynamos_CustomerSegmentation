```python

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# Step 1: Load the dataset
from google.colab import files
uploaded = files.upload()  # Upload the 'Online Retail.xlsx' file

data = pd.read_excel("2.xlsx")

# Step 2: Data Cleaning
data.dropna(subset=['CustomerID'], inplace=True)  # Remove rows without CustomerID
data['TotalPrice'] = data['Quantity'] * data['UnitPrice']  # Calculate total price per transaction

# Step 3: Feature Engineering
customer_data = data.groupby('CustomerID').agg(
    Total_Spending=('TotalPrice', 'sum'),
    Num_Transactions=('InvoiceNo', 'nunique'),
    Avg_Spending=('TotalPrice', 'mean')
).reset_index()

# Step 4: Simple Analysis (No Clustering)
# Highlight the top 10 customers based on Total Spending
top_customers = customer_data.nlargest(10, 'Total_Spending')

# Step 5: Visualization of Spending Patterns
plt.figure(figsize=(12, 8))

# Scatter plot of all customers
plt.scatter(customer_data['Num_Transactions'], customer_data['Total_Spending'], alpha=0.5, label='All Customers')

# Highlight the top customers
plt.scatter(top_customers['Num_Transactions'], top_customers['Total_Spending'], color='red', label='Top 10 Customers')

# Add labels and title
plt.xlabel("Number of Transactions")
plt.ylabel("Total Spending")
plt.title("Customer Spending Patterns")
plt.legend()
plt.show()

# Bar plot of the top customers
plt.figure(figsize=(10, 6))
plt.bar(top_customers['CustomerID'].astype(str), top_customers['Total_Spending'], color='orange')
plt.xlabel("Customer ID")
plt.ylabel("Total Spending")
plt.title("Top 10 Customers by Total Spending")
plt.xticks(rotation=45)
plt.show()
```
